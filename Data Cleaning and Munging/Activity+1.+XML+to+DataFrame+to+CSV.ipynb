*Activity 1: Transform XML to data frame to CSV*

*Washington University data file*


- Note:
 1. Used XMLtoList to convert the XML to list
 2. Used ldply function to unlist the list and store it in a dataframe.. Doing this because every node can have any number of children nodes
 3. Used write.csv to store the dataframe in CSV
 

- Installing and Importing Libraries required

```r
install.packages("XML", repos = "http://cran.us.r-project.org")
```

- Include liraries XML and plyr
 - XML: for using xmlToList
 - plyr: for using ldply

```r
library(XML)
library(plyr)
```

- Converting the XML to List 

```r
dataList <- xmlToList("reed.xml", addAttributes = TRUE)
```

```r
dataList
```

```r
data.frame(t(unlist(dataList)))
```

- Converting the list to dataframe using ldply and recursion function

```r
dataFrame <- ldply(dataList,function(s){
    data.frame(t(unlist(s)))
},.id=NULL)
```

- Write the dataframe to CSV

```r
write.csv(dataFrame, file = "reed.csv",row.names=FALSE)
```
